```python
import lyricsgenius
import pandas as pd
import re
import time
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 500)
```

```python
# Replace 'your_access_token' with your actual Genius API token
genius = lyricsgenius.Genius("RY1wLVIh1PXpg98DXGOfFdVbKRtY7kM_7UnFBdDfEsfhQOrVb1DqDxjocYJzmQPQ", timeout=15)  # Increase the timeout to 15 seconds
```

```python
# The name of the artist and the albums to fetch
artist_name = "Justin Bieber"
albums = ["Purpose", "Justice"]

# Initialize a list to store song data
song_data = []
```

```python
# Function to clean the lyrics by removing unwanted text using regex
def clean_lyrics(lyrics):
    # Remove unwanted metadata using regex
    lyrics = re.sub(r'\d+\sContributors.*?Translations.*?\n', '', lyrics, flags=re.DOTALL)
    lyrics = re.sub(r'\d+\sContributors.*?\n', '', lyrics, flags=re.DOTALL)
    
    # Remove '[number]Embed' at the end of the lyrics or any '[Produced by...]'
    lyrics = re.sub(r'\[.*?Embed\]', '', lyrics, flags=re.IGNORECASE)
    lyrics = re.sub(r'\[Produced by.*?\]', '', lyrics, flags=re.IGNORECASE)
    
    # Specifically target 'numberEmbed' at the end of the lyrics
    lyrics = re.sub(r'\d+Embed$', '', lyrics, flags=re.IGNORECASE)
    
    # Remove labels like [Chorus], [Verse], [Bridge], [Pre-Chorus], etc.
    lyrics = re.sub(r'\[.*?\]', '', lyrics, flags=re.IGNORECASE)
    
    # Further clean up any leftover newlines or extra spaces
    lyrics = lyrics.strip()
    
    return lyrics
```

```python
# Loop through each album and fetch songs by album
for album_name in albums:
    print(f"Fetching songs from album: {album_name}")
    try:
        album = genius.search_album(album_name, artist_name)
        for track in album.tracks:
            song_info = track.song  # Access the song object from the Track object
            song_title = song_info.title  # Get the song title from the song object
            try:
                lyrics = clean_lyrics(song_info.lyrics)
                song_data.append({
                    "title": song_title,
                    "artist": artist_name,
                    "album": album_name,
                    "lyrics": lyrics
                })
            except Exception as e:
                print(f"Error fetching song '{song_title}': {e}")
                time.sleep(5)  # Wait 5 seconds before retrying
    except Exception as e:
        print(f"Error fetching album '{album_name}': {e}")
        time.sleep(5)  # Wait 5 seconds before retrying
```

```python
# Convert the list to a DataFrame
df = pd.DataFrame(song_data)

# Save the DataFrame to a CSV file
csv_file_path = 'justin_bieber_lyrics_albums_cleaned.csv'
df.to_csv(csv_file_path, index=False)

print(f"Lyrics from the albums saved to {csv_file_path}")
```
